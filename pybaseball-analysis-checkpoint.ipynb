# Statcast

Statcast data include pitch-level features such as Perceived Velocity (PV), Spin Rate (SR), Exit Velocity (EV), pitch X, Y, and Z coordinates, and more. The function statcast(start_dt, end_dt) pulls this data from baseballsavant.com.

```python
statcastdf = statcast(start_dt=str(today-timedelta(days=7)), end_dt=str(today))
```

```python
print(statcastdf.head())
```

# Batting Stats - By Range

```python
data = batting_stats_range(start_dt=str(today-timedelta(days=365)), end_dt=str(today))
```

```python
data.columns
```

```python
data.head()
```

# Batting Stats EDA

https://plotly.com/python/plotly-express/

```python
import plotly.express as px
```

Examples: Age isn't everything!

```python
fig = px.scatter(data, x="Age", y="BA", color="Lev", marginal_y="violin",
           marginal_x="violin", trendline="ols", template="simple_white", title="Batting Average by Age")
fig.show()
```

```python
fig = px.histogram(data, x='Age', y='BA', barmode='group', histfunc='avg', color='Lev', title="Batting Average by Age")
fig.show()
```

```python
data_grouped = data.groupby(['Age', 'Lev'], as_index=False)['BA'].mean()
data_grouped
```

```python
fig = px.line(data_grouped, x='Age', y='BA', color='Lev', title="Batting Average by Age")
fig.show()
```
